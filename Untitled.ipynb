## Creating and Understandig Data

```python
# Importing relevant libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Importing data and looking at the sample data
df = pd.read_csv("day.csv", index_col= 'instant' )
df.head()
```

```python
# Getting Details of the data
df.info()
# Clearly we have all non- null values
```

## Data Visualization

```python
# Droping unnecessary columns
df.drop(["dteday","casual","registered"], axis= 1, inplace= True)
df.head()
```

```python
# Mappig Categorical variables with actual values
season =  {1:"spring", 2:"summer", 3:"fall", 4:"winter"}
#yr = {0: '2018', 1:'2019'}
mnth ={1: "Jan",2: "Feb",3:"Mar", 4: "Apr",5: "May",6: "Jun",7: "Jul",8: "Aug",9: "Sep",10:"Oct" ,11:"Nov",12:"Dec"}
holiday= {0:"No",1: "Yes"}
weekday ={0:"Mon",1:"Tue",2:"Wed",3:"Thu",4:"Fri",5:"Sat",6:"Sun"}
workingday= {0:"No",1: "Yes"}
weathersit = {1:"1",2:"2",3:"3",4:"4"}
```

```python
df['season'] = df['season'].map(season)
#df['yr'] = df['yr'].map(yr)
df['mnth'] = df['mnth'].map(mnth)
df['holiday'] = df['holiday'].map(holiday)
df['weekday'] = df['weekday'].map(weekday)
df['workingday'] = df['workingday'].map(workingday)
df['weathersit'] = df['weathersit'].map(weathersit)
df
```

```python
# Performing EDA on the data
sns.pairplot(df)
```

```python
df.columns
```

```python
# Checking the distribution of numerical variable 
#(since there are only 730 rows and there are not major outliers we are not planning to remove them)
df_cols_num = ['temp', 'atemp', 'hum', 'windspeed', 'cnt']
for i in df_cols_num:
    sns.boxplot(y =df[i])
    plt.show()
```

```python
# Checking Correlation between variables
plt.figure(figsize=(10,10)) 
sns.heatmap(df.corr(),annot= True,cmap="YlGnBu")
```

## Data Preparation

```python
df.drop('atemp',axis = 1, inplace = True)
df
```

```python
# Performing Dummy Encoding of the data
df = pd.get_dummies(df, drop_first = True)
```

```python
# Checking Correlation between variables
plt.figure(figsize=(20,20)) 
sns.heatmap(df.corr(),annot= True,cmap="YlGnBu")
```

```python
df_train, df_test = train_test_split(df, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
df_train.head()
```

```python
scaler = MinMaxScaler()
df_cols_num = ['temp', 'hum', 'windspeed', 'cnt']
df_train[df_cols_num] = scaler.fit_transform(df_train[df_cols_num])
```

```python
y_train = df_train.pop('cnt')
X_train = df_train
```

```python
X_train
```

## Model Building

```python
# Adding constant to the X_train
X_train_lm = sm.add_constant(X_train)
lr_model = sm.OLS(y_train, X_train_lm).fit()
print(lr_model.params)
lr_model.summary()
```

```python
### Using Recursive feature elimination to find only top x variables
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
rfe = RFE(lm,n_features_to_select = 15,step = 1)        
rfe = rfe.fit(X_train, y_train)
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

##### Re-running model using only top 15 features

```python
X_train_lm = X_train[col]
```

```python
X_train_lm.columns
```

```python
# Adding constants 
X_train_rfe = sm.add_constant(X_train_lm)
X_train_rfe
```

```python
lr_model = sm.OLS(y_train, X_train_rfe).fit()
print(lr_model.params)
lr_model.summary()
```

```python
vif = pd.DataFrame()
vif['Columns'] = X_train_lm.columns
vif['VIF'] = [variance_inflation_factor(X_train_lm.values, i) for i in range(len(X_train_lm.columns))]
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
X_train_lm.drop(['hum','temp'], axis=1, inplace = True)
```

```python
X_train_lm
```

```python
# Adding constants 
X_train_rfe = sm.add_constant(X_train_lm)
X_train_rfe
```

```python
lr_model = sm.OLS(y_train, X_train_rfe).fit()
print(lr_model.params)
lr_model.summary()
```

```python
vif = pd.DataFrame()
vif['Columns'] = X_train_lm.columns
vif['VIF'] = [variance_inflation_factor(X_train_lm.values, i) for i in range(len(X_train_lm.columns))]
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
# Now both VIF and R2 are in range 
```
